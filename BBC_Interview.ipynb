```python
#6.	Write a Pandas program to find out the alcohol consumption of a given year from the world alcohol consumption dataset

alcohol.groupby('Year').sum()
```

```python
#7.	Write a Pandas program to find out the alcohol consumption details in the year '1987' or '1989' from the world alcohol consumption dataset

alcohol[(alcohol['Year'] == 1987) | (alcohol['Year'] == 1989)].sort_values(by= 'Year')
```

```python
#8.	Write a Pandas program to find out the alcohol consumption details by the 'Americas' in the year '1985' from the world alcohol consumption dataset

alcohol [(alcohol['WHO region'] == 'Americas') & (alcohol['Year'] == 1985)]
```

```python
#9.	Write a Pandas program to find out the alcohol consumption details in the year '1986' where WHO region is 'Western Pacific' and country is 'VietNam'

alcohol[(alcohol['Year'] == 1986  & (alcohol['WHO region'] == 'Western Pacific') & (alcohol['Country'] == 'Viet Nam'))]
```

```python
#10.	Write a Pandas program to find out the alcohol consumption details in the year '1986' or '1989' where WHO region is 'Americas' from the world alcohol 
#consumption dataset.



alcohol[((alcohol['Year']== 1986) | (alcohol['Year']== 1989))& ((alcohol['WHO region'] == 'Americas'))]
```

```python
#Write a Pandas program to find out the alcohol consumption details in the year '1986' or '1989' where WHO region is 'Americas' or 'Europe' 
#from the world alcohol consumption dataset

alcohol[((alcohol['Year'] == 1986) | (alcohol['Year'] == 1989)) & ((alcohol['WHO region'] == 'Americas') | (alcohol['WHO region'] == 'Europe'))]
```

```python
#Alternatively

alcohol[(alcohol['WHO region'].isin(['Americas', 'Europe']))& (alcohol['Year'].isin([1986, 1989])) ]
```

```python
#Write a Pandas program to find out the 'WHO region, 'Country', 'Beverage Types' in the year '1986' or '1989' 
#where WHO region is 'Americas' or 'Europe' from the world alcohol consumption dataset.

newYear = alcohol[(alcohol['Year'] == 1986 )| (alcohol['Year'] == 1989)]
newYear
```

```python
newYear[['WHO region', 'Country', 'Beverage Types']].head()
```
